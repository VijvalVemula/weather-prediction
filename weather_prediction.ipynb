```python
from sklearn.linear_model import Ridge
```

```python
def backtest(data, model, predictors, start = 3650, step = 90):
    all_predictions = [] 
    for i in range(start, data.shape[0], step):
        train = data.iloc[ : i, : ]
        test = data.iloc[i : (i + step), : ]
        model.fit(train[predictors], train["target"])
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index = test.index)
        combined = pd.concat([test["target"], preds], axis = 1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["actual"] - combined["prediction"]).abs()
        all_predictions.append(combined)
    return pd.concat(all_predictions, axis = 0)
```

```python
predictions = backtest(data, rr, predictors)
predictions
```

```python
from sklearn.metrics import mean_absolute_error


mean_absolute_error(predictions["actual"], predictions["prediction"]) # use this method since it converts the differences into absolute values behind the screen
```

```python
"""def pct_diff(old, new):
    return 100 * ((new - old) / old)"""

pct_diff = lambda old, new : 100 * ((new - old) / old)

def compute_rolling(data, horizon, col):
    label = f"rolling_{horizon}_{col}"
    data[label] = data[col].rolling(horizon).mean()
    data[f"{label}_pct"] = pct_diff(data[label], data[col])
    return data
```

```python
rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        data = compute_rolling(data, horizon, col)
```
